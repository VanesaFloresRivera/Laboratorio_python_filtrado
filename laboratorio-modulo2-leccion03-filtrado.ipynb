```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', None) 
import numpy as np
```

```python
# Carga el dataset de trabajo
df_netflix_agrupado = pd.read_csv("datos/Netflix_full.csv", index_col=[0])
df_netflix_agrupado.head(10)
```

#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
df_netflix_agrupado.info()
```

```python
filtro_tipo_movie = df_netflix_agrupado.type == 'Movie'
filtro_año_lanzamiento = df_netflix_agrupado.release_year == 2021
df_movie_2021 = df_netflix_agrupado[filtro_tipo_movie & filtro_año_lanzamiento]
df_movie_2021
```

```python
#comprobación filtro
print(df_movie_2021.type.unique())
print(df_movie_2021.release_year.unique())
```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
filtro_duracion_90 = df_netflix_agrupado.duration == '90 min'
filtro_runtime_90 = df_netflix_agrupado.Runtime == 90

df_filtro_duracion = df_netflix_agrupado[filtro_duracion_90|filtro_runtime_90]
df_filtro_duracion.loc[:,['title', 'duration','Runtime' ]]
```

```python
#comprobación filtro
print(df_filtro_duracion.duration.unique())
print(df_filtro_duracion.Runtime.unique())
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos a dirigido cada uno de los directores?

```python
filtro_director = ['Christopher Nolan', 'Martin Scorsese']
df_director = df_netflix_agrupado[df_netflix_agrupado.director.isin(filtro_director)]
df_director
```

```python
#¿Cuantos ha dirigido cada uno de los directores?
df_contador_direcciones = pd.DataFrame(df_director.director.value_counts())
display(df_contador_direcciones)

print(f'El director {df_contador_direcciones.index[0]}, ha dirigido {df_contador_direcciones.iloc[0,0]} películas ')

print(f'El director {df_contador_direcciones.index[1]}, ha dirigido {df_contador_direcciones.iloc[1,0]} películas ')
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python
filtro_paises = ['United States', 'United Kingdom', 'Canada']
df_varios_paises = df_netflix_agrupado[df_netflix_agrupado.country.isin(filtro_paises)]
print(f'Los títulos que pertenencen a los países United States, United Kingdom y Canada son:')
df_varios_paises
```

#### Ejercicio 7: ¿Qué títulos fueron lanzados entre 2015 y 2020, inclusive?

```python
filtro_año_lanzamiento_inicio= df_netflix_agrupado.release_year >= 2015
filtro_año_lanzamiento_final = df_netflix_agrupado.release_year <= 2020

df_lanzamientos_2015_2020= df_netflix_agrupado[filtro_año_lanzamiento_inicio & filtro_año_lanzamiento_final]
df_lanzamientos_2015_2020
```

```python
#comprobación filtro
print(df_lanzamientos_2015_2020.release_year.unique())
```

```python
#Con el método between:
df_lanzamientos_2015_2020_1 = df_netflix_agrupado[df_netflix_agrupado.release_year.between(2015,2020, inclusive='both')]
df_lanzamientos_2015_2020_1
```

```python
#comprobación filtro
print(df_lanzamientos_2015_2020_1 .release_year.unique())
```

#### Ejercicio 8: ¿Cuáles son los títulos cuya puntuación en IMDb está entre 6.5 y 8.0 y que no sepamos cual es su duración?

```python
df_puntuacion= df_netflix_agrupado[df_netflix_agrupado['IMDB Score'].between(6.5,8, inclusive='both')]
print('Los títulos con una puntuación entre 6.5 y 8 son:')
display(df_puntuacion)
print(f'Los títulos dentro de esta puntuación de los que no conozcamos su duración son:')
df_puntuacion_sinduracion = df_puntuacion[(df_netflix_agrupado.duration.isnull())&(df_netflix_agrupado.Runtime.isnull())]
display(df_puntuacion_sinduracion)

if df_puntuacion_sinduracion.shape[0] == 0:
    print(f'Todos los títulos tienen duración')
```
